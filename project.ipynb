```python
#check crs and reproject layers
def reproject_to_epsg25832(gdf):
    # Check the CRS of the input GeoDataFrame
    print('Original CRS:', gdf.crs)

    # Reproject the GeoDataFrame to EPSG 25832
    gdf_25832 = gdf.to_crs(epsg=25832)

    # Check the CRS of the output GeoDataFrame
    print('Reprojected CRS:', gdf_25832.crs)

    return gdf_25832

# apply function
admin_pop_25832 = reproject_to_epsg25832(admin_pop)
noise_25832 = reproject_to_epsg25832(noise)
housing_25832 = reproject_to_epsg25832(housing)

```

```python
def intersect_and_calculate_area(housing_25832, noise_25832):
    # Perform spatial intersection of housing layer with noise layer
    intersection = gpd.overlay(housing_25832, noise_25832, how='intersection')
    
    # Group intersection by noise class and administrative boundary
    intersection_grouped = intersection.groupby(['klasse'])['geometry'].agg(lambda x: x.area).reset_index()
    
    # Rename area column to be more descriptive
    intersection_grouped = intersection_grouped.rename(columns={'geometry': 'intersected_area'})
    
    # Return the grouped and calculated intersection
    return intersection_grouped
```
